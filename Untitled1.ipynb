```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SampleSuperstore.csv')
```

```python
# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())

# Check for missing values and data types
print("\nInformation about the dataset:")
print(df.info())

# Descriptive statistics of the dataset
print("\nDescriptive statistics of the dataset:")
print(df.describe())

# Check for duplicate rows
print("\nNumber of duplicate rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python

# Correlation matrix
numeric_df = df.select_dtypes(include='number')
corr = numeric_df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Univariate Analysis: Distribution of Profit
plt.figure(figsize=(8, 6))
sns.histplot(df['Profit'], bins=30, kde=True)
plt.title('Distribution of Profit')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.show()

```

```python
# Bivariate Analysis: Sales vs Profit
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
# Select only numeric columns
numeric_columns = df.select_dtypes(include=np.number).columns

# Segmenting data by Category and calculating mean for numeric columns
category_group = df.groupby('Category')[numeric_columns].mean()
print("\nMean values by Category:")
print(category_group)
```

```python
# Visualizing Sales, Quantity, Discount, and Profit by Category
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

sns.barplot(x='Category', y='Sales', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Sales by Category')

sns.barplot(x='Category', y='Quantity', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Quantity by Category')

sns.barplot(x='Category', y='Discount', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Discount by Category')

sns.barplot(x='Category', y='Profit', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Profit by Category')

plt.tight_layout()
plt.show()
```

```python
# Select only numeric columns
numeric_columns = df.select_dtypes(include=np.number).columns

# Segmenting data by State and calculating mean for numeric columns
state_group = df.groupby('State')[numeric_columns].mean()
print("\nMean values by State:")
print(state_group)


```

```python

# Visualizing Sales and Profit by State
plt.figure(figsize=(20, 10))
sns.barplot(x='State', y='Sales', data=df)
plt.title('Sales by State')
plt.xlabel('State')
plt.ylabel('Sales')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(20, 10))
sns.barplot(x='State', y='Profit', data=df)
plt.title('Profit by State')
plt.xlabel('State')
plt.ylabel('Profit')
plt.xticks(rotation=90)
plt.show()
```

```python
# Segmenting data by Segment
# Segmenting data by Segment
numeric_columns = df.select_dtypes(include=np.number).columns
segment_group = df.groupby('Segment')[numeric_columns].mean()
print("\nMean values by Segment:")
print(segment_group)

```

```python

# Visualizing Sales and Profit by Segment
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.barplot(x='Segment', y='Sales', data=df, ax=axes[0])
axes[0].set_title('Sales by Segment')

sns.barplot(x='Segment', y='Profit', data=df, ax=axes[1])
axes[1].set_title('Profit by Segment')

plt.tight_layout()
plt.show()
```

```python

# Identifying Weak Areas for Improvement
# Example: High Discount leading to Low Profit
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()

# Example: Low Sales and High Profit Categories
low_sales_high_profit = df[(df['Sales'] < df['Sales'].mean()) & (df['Profit'] > df['Profit'].mean())]
print("\nLow Sales, High Profit Categories:")
print(low_sales_high_profit[['Category', 'Sub-Category', 'Sales', 'Profit']])

# Example: High Sales and Low Profit Categories
high_sales_low_profit = df[(df['Sales'] > df['Sales'].mean()) & (df['Profit'] < df['Profit'].mean())]
print("\nHigh Sales, Low Profit Categories:")
print(high_sales_low_profit[['Category', 'Sub-Category', 'Sales', 'Profit']])
```
